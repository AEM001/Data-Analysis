（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
import numpy as np 
import matplotlib.pyplot as plt
x=np.linspace(-5,3)
y=x**2+2*x+1
plt.plot(x,y)
plt.show()
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
plt.subplot(211)
t=np.linspace(-3.14,3.14)
plt.plot(t,np.sin(t),color='b',linewidth=2.5)
plt.subplot(212)
plt.plot(t,np.cos(t),color='r',linestyle='--')
```

（3）读取上次作业保存的酒店数据，画出每个地区酒店数量的柱状图，柱状颜色为红色

```python
import pandas as pd
df=pd.read_excel("../作业5/酒店数据2.xlsx")
data=df['地区'].value_counts()
data
```

```python
x=data.index
y=data.values
plt.figure(figsize=(10,6))
plt.bar(x,y,color='b')
plt.title("the number of hotels each area have",fontsize=20)
plt.xlabel("area")
plt.ylabel("number of hotels")
plt.tick_params(labelsize=14)
plt.xticks(rotation=90)
for a,b in zip(x,y):
    plt.text(a,b+10,b,ha="center",va="bottom",fontsize=10)
plt.show()
```

（4）画出每个价格等级酒店数量的柱状图。

```python
data=df['价格等级'].value_counts()
x=data.index
y=data.values
plt.figure(figsize=(8,8))
plt.bar(x,y)
```

（5）画出各个价格等级占比的饼图。

```python
plt.figure(figsize=(8,6))
plt.title("pie chart")
patches,l_text,p_text=plt.pie(y,labels=x,autopct="%.1f%%",colors="brg",startangle=45)
for i in p_text:
    i.set_size(15)
    i.set_color("w")

for i in l_text:
    i.set_size(15)
    i.set_color("k")

plt.legend()
plt.show()
```

（6）画出酒店评分的直方图。

```python
plt.figure(figsize=(10,8))
plt.hist(df['评分'],bins=50,edgecolor='k')
plt.show()
```

（7）画出每个热门等级酒店评分均值的柱状图。（按照评分均值从小到大排序。）

```python
mean_score=df.groupby('热门等级')['评分'].mean()
mean_score
plt.figure(figsize=(4,6))
plt.bar(mean_score.index,mean_score,color='b')
for a,b in zip(mean_score.index,mean_score):
    plt.text(a,b+0.05,f"{b:.2f}",ha="center",va="bottom",fontsize=10)
```
