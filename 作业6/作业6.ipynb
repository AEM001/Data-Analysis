（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
import numpy as np 
import matplotlib.pyplot as plt
x=np.linspace(-5,3)
y=x**2+2*x+1
plt.plot(x,y)
plt.show()
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
plt.subplot(211)
t=np.linspace(-3.14,3.14)
plt.plot(t,np.sin(t),color='b',linewidth=2.5)
plt.subplot(212)
plt.plot(t,np.cos(t),color='r',linestyle='--')
```
