（1）L=[1,2,'','my',3,'name','is',4,'katty'],利用循环语句和判断条件，分别输出列表中的字符串和数字。

```python
L=[1,2,'','my',3,'name','is',4,'katty']
str=list()
num=list()
print(type(L[2]))
for item in L:
    if type(item)==str:
        str.append(item)
    else:
        num.append(item)
print(str)
print(num)
```

（3）定义一个判断字符串的长度是否大于10的函数。

```python
def len_over_10(a: str) -> bool:
    if len(a) > 10:
        return True
    else:
        return False
len_over_10('hello world')
```

（4）定义一个求解阶乘的函数。

```python
def jiec(a:int)->int:
    n=a
    while n>1:
        n-=1
        a*=n
    return a
jiec(5)
```

（5）利用列表生成式，生成1-5的阶乘。

```python
print([jiec(i) for i in range(1,6)])
```

（6）利用函数和列表生成式，标记一个列表，奇数标记为1，偶数标记为2，并且统计一下奇数和偶数的数量。
     例如：[1,4,2,4,2,9,5]，得到[1,2,2,2,2,1,1]

```python
def is_odd(a:int)->int:
    if a%2==1:
        return 1
    else:
        return 2
l=[1,4,2,4,2,9,5]
list_m=[is_odd(i) for i in l]
print(list_m)
print(f"the num of odd is {list_m.count(1)}")
print(f"the num of even is {list_m.count(2)}")
```
