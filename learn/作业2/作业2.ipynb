（10）把一个10\*2的随机生成的笛卡尔坐标转换成极坐标。

```python
a10=np.random.rand(10,2)
x=a10[:,0]
y=a10[:,1]

r=np.sqrt(x**2+y**2)
theta=np.arctan2(y,x)

# polar_a10=np.dstack((r,theta))
polar_a10=np.column_stack((r,theta))
print(polar_a10)
```

（14）构造两个2乘3的二维数组，按照3种方法进行连接？

```python
a141=np.ones((2,3))
a142=np.ones((2,3))*2
a14=np.concatenate((a141,a142),axis=1)
print(a14)
a14=np.concatenate((a141,a142),axis=0)
print(a14)
a14=np.vstack((a141,a142))
print(a14)
a14=np.hstack((a141,a142))
print(a14)
a14=np.dstack((a141,a142))
print(a14)
```

（15）获取数组 a 和 b 中的共同项（索引位置相同，值也相同）。
a = np.array([1,2,3,2,3,4,3,4,5,6])，b = np.array([7,2,10,2,7,4,9,4,9,8])

```python
a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])
aidx=np.where(a==b)
print("aidx:",aidx)
print(a[aidx])
```
