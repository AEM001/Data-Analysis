（12）用数据透视表，计算每个类型的酒店的评分人数总数量。

```python
df.columns
```

```python
# 用数据透视表，计算每个类型的酒店的评分人数总数量
pivot_table_ratings = pd.pivot_table(df, index=['地区'], values=['评分','价格','评分人数'], aggfunc={'评分人数':'sum','价格':'mean','评分':'mean'})
# print("原始数据透视表:")
# print(pivot_table_ratings)
print("\n排序:")
print(pivot_table_ratings.sort_values(by=['评分人数','评分'], ascending=[False,False]))
```

```python
pd.pivot_table(df,index=["地区","类型"],values='价格')
```

（13）用数据透视表，计算每个类型的酒店价格的均值和标准差
