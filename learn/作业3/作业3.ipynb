## （2）数据操作

（1）读取香港酒店数据。

```python
df=pd.read_excel("香港酒店数据.xlsx",index_col=0)
```

（2）按照数据的内容，重新设置数据的索引，重新设置列名称为'名字','类型','城市','地区','地点','评分','评分人数','价格'。

```python
df.head()
```

```python
df.index=list(range(1,len(df)+1))
```

```python
df.head()
```

```python
df.drop(1, axis=0, inplace=True)
df.index=range(1, len(df) + 1)
df.head()
```

```python
df.columns=['名字','类型','城市','地区','地点','评分','评分人数','价格']
```

（3）查看所有类型为“浪漫情侣”的酒店

```python
love=df[df["类型"]=="浪漫情侣"]
```

（4）查看所有类型为“浪漫情侣”，地区在湾仔的酒店

```python
bayboy=love[love["地区"]=="湾仔"]
bayboy
```

（5）查看所有地址在观塘或者油尖旺，评分大于4的酒店

```python
# 错误1：使用了位运算符，应使用逻辑运算符
# 错误2：使用了双等号，应使用单等号
# 错误3：使用了英文字符串，应使用中文字符串
# 正确的代码为：
df[(df['地区'].isin(['观塘', '油尖旺'])) & (df['评分'] > 4)]
```
