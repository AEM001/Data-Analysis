（1）读取酒店数据2.xlsx，根据评分和价格信息，绘制散点图。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_excel('../作业6/酒店数据2.xlsx')
x=df['评分']
y=df['价格']
print(y.sort_values(ascending=True))
# y.drop(y.sort_values(ascending=True).index[-1], inplace=True)
index=y.sort_values(ascending=True).index[-1]
x.pop(index)
y.pop(index)
plt.figure(figsize=(8,6))
plt.scatter(x,y,color='c',marker='p',label='price')
plt.legend()
plt.title("score and price scatter plot")
plt.xlabel('score')
plt.ylabel('price')
plt.show()
```

pop(label)Removes and returns the element at the given index label.works in-place
drop(labels) Removes the element(s) at the given index label(s). returns a new series
