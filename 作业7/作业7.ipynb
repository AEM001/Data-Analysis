（1）读取酒店数据2.xlsx，根据评分和价格信息，绘制散点图。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_excel('../作业6/酒店数据2.xlsx')
print(df['价格'].max())
x=df['评分']
y=df['价格']
print(y.sort_values(ascending=True))
# y.drop(y.sort_values(ascending=True).index[-1], inplace=True)
# index=y.sort_values(ascending=True).index[-1]
# x.pop(index)
# y.pop(index)
plt.figure(figsize=(8,6))
plt.scatter(x,y,color='c',marker='p',label='price')
plt.legend()
plt.title("score and price scatter plot")
plt.xlabel('score')
plt.ylabel('price')
plt.show()
print(df['价格'].max())
```

pop(label)Removes and returns the element at the given index label.works in-place
drop(labels) Removes the element(s) at the given index label(s). returns a new series

（2）画出油尖旺地区，评分的箱线图。

```python
data=df[df.地区=='油尖旺']['评分']
plt.figure(figsize=(8,6))
plt.boxplot(
    data,
    whis=2,
    flierprops={'marker':'o', 'markerfacecolor':'r', 'color':'k'},
    patch_artist=True,
    boxprops={'color':'k', 'facecolor':'#66ccff'},vert=True
)
plt.title("boxplot of score in 油尖旺",fontsize=10)
plt.show()
```

（3）选出平均价格前5的地区，画出这些地区的评分的箱线图。

```python
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] # macOS 自带的通用中文字体
plt.rcParams['axes.unicode_minus'] = False # 解决负号 '-' 显示为方块的问题
ave_price=df.groupby('地区')['价格'].mean()#是一个series
ave_price=ave_price.sort_values(ascending=False)[:5]
score_data=[df.loc[df['地区']==area]['评分'] for area in ave_price.index]
plt.figure(figsize=(8,6))
# plt.boxplot(df.loc[ave_price.index]['评分'],labels=ave_price.index,whis=1.5,flierprops={'marker':'o',"markerfacecolor":'r',"color":"k"}
#             ,patch_artist=True,boxprops={"color":'k',"facecolor":'#99ff99'},vert=False)

plt.boxplot(score_data,labels=ave_price.index,whis=1.5,flierprops={'marker':'o',"markerfacecolor":'r',"color":"k"}
            ,patch_artist=True,boxprops={"color":'k',"facecolor":'#99ff99'},vert=True)
ax=plt.gca()
ax.patch.set_facecolor('gray')
ax.patch.set_alpha(0.3)
plt.title("box plot for top5 average price area",fontsize=10)
plt.show()
```
